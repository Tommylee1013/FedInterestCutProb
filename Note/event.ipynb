```python
def find_prob_soar_dates(
        prob_data : pd.DataFrame,
        col_name : str = 'prob_0',
        fomc_dates : pd.DatetimeIndex = fomc_dates,
        threshold : float = 0.4
    ) -> pd.DatetimeIndex :

    diff_data = (prob_data[col_name].diff() >= threshold).astype(int).loc[
        list(set(prob_data.index).difference(pd.to_datetime(fomc_dates) + pd.Timedelta(days = 1)))
    ].sort_index()

    res = diff_data[diff_data == 1].index

    return res
```

```python
interest_freeze_prob_soar_event = find_prob_soar_dates(
    prob_data = prob_data.loc['2010':],
    col_name = 'prob_0',
    fomc_dates = fomc_dates,
)

interest_25cut_prob_soar_event = find_prob_soar_dates(
    prob_data = prob_data.loc['2010':],
    col_name = 'prob_-25',
    fomc_dates = fomc_dates,
)

interest_25hike_prob_soar_event = find_prob_soar_dates(
    prob_data = prob_data.loc['2010':],
    col_name = 'prob_25',
    fomc_dates = fomc_dates,
)

interest_50cut_prob_soar_event = find_prob_soar_dates(
    prob_data = prob_data.loc['2010':],
    col_name = 'prob_-50',
    fomc_dates = fomc_dates,
)

interest_50hike_prob_soar_event = find_prob_soar_dates(
    prob_data = prob_data.loc['2010':],
    col_name = 'prob_50',
    fomc_dates = fomc_dates,
)

interest_75cut_prob_soar_event = find_prob_soar_dates(
    prob_data = prob_data.loc['2010':],
    col_name = 'prob_-75',
    fomc_dates = fomc_dates,
)

interest_75hike_prob_soar_event = find_prob_soar_dates(
    prob_data = prob_data.loc['2010':],
    col_name = 'prob_75',
    fomc_dates = fomc_dates,
)
```
